#Avaliando o tamanho Arquivos

```python
#Verficando o tamanho de cada arquivo

#função para calcular o tamanho do arquivo
#codigo obtido no stackoverflow.com
def valor_bytes(num):
    #This function will covert bytes to MB ... GB ... etc
    for x in ['bytes', 'KB', 'MB', 'GB', 'TB']:
        if num <1024.0:
            return "%3.1f %s" % (num, x)
        num /= 1024.0

cv_bytes_abr = os.path.getsize(
    path_abr
)

print("Tamanho do arquivo Abril é de: ", valor_bytes(cv_bytes_abr))

# cv_bytes_fev = os.path.getsize(
#     path_fev
# )

# print("Tamanho do arquivo Fevereiro é de: ", valor_bytes(cv_bytes_fev))

# cv_bytes_mar = os.path.getsize(
#     path_mar
# )

# print("Tamanho do arquivo Março é de: ", valor_bytes(cv_bytes_mar))

# cv_bytes_mar2 = os.path.getsize(
#     path_mar2
# )

# print("Tamanho do arquivo Dezembro é de: ", valor_bytes(cv_bytes_mar2))

```
